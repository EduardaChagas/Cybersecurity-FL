```python
def convert_back(img):
    c_std = [0.247, 0.243, 0.261]
    c_mean = [0.4914, 0.4822, 0.4466]
    img[0] = img[0] * c_std[0] + c_mean[0]
    img[1] = img[1] * c_std[1] + c_mean[1]
    img[2] = img[2] * c_std[2] + c_mean[2]
    #img = img.reshape(32, 32, 3)
    img = img.permute(1, 2, 0)
    return img
```
